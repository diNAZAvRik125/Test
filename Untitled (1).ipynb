```python
import pandas as pd
import numpy as np
import pylab as plt
import seaborn as sns
```

```python
df = df.loc[df['fact'] != 'NaN']
print(df.head)
```

```python
df.columns
```

```python
sort_by_year = df.sort_values('year')
```

```python
selected_columns = ['Город', 'year', 'Модель', 'Нижняя граница', 'Верхняя граница']
data = sort_by_year[selected_columns]
f, ax = plt.subplots(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr,
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            vmin=-1.0, vmax=1.0,
            square=True, ax=ax)

plt.show()
```

### Просмотр зависимостей на всем датасете не дало результатов, зависимости исследовать не вышло. Взглянем на графики зависимостей различных параметрво по годам ###

```python
grap = sort_by_year.groupby('year')['Модель'].mean()
plt.plot(grap.index, grap.values)
plt.xlabel('year')
plt.ylabel('Модель')
plt.title('Average year per Модель')
plt.show()
```

### Если бы достоверно знали, что такое "Модель" то могли бы точно сказать что что-то растет судя по годам ###

```python
grap = sort_by_year.groupby('year')['Нижняя граница'].mean()
plt.plot(grap.index, grap.values)
plt.xlabel('year')
plt.ylabel('Нижняя граница')
plt.title('Average year per Нижняя граница')
plt.show()
```

### Из описания задачи предположу, что "Нижняя граница" - это нижняя граница населения. Судя по данным она растет и в 2120 году достигнет пика  ###

```python
grap = sort_by_year.groupby('year')['Верхняя граница'].mean()
plt.plot(grap.index, grap.values)
plt.xlabel('year')
plt.ylabel('Верхняя граница')
plt.title('Average year per Верхняя граница')
plt.show()
```

```python
import plotly.express as px
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.layouts import column
```

```python
source = ColumnDataSource(df)
output_file("dashboard.html")
p1 = figure(x_range=df['Город'], title="Фактическое и прогнозное население с доверительными интервалами", plot_height=400, toolbar_location=None)
p1.vbar(x='Город', top='fact', width=0.4, source=source, legend_label="Фактическое", color="blue", alpha=0.7)
p1.vbar(x='Город', top='Модель', width=0.4, source=source, legend_label="Модель", color="orange", alpha=0.6)
p1.segment(x0='Город', y0='Нижняя граница', x1='Город', y1='Верхняя граница', source=source, line_color="green", line_width=2, legend_label="Доверительный интервал")
p1.xaxis.major_label_orientation = "vertical"
p1.yaxis.axis_label = "Численность населения"
p1.legend.location = "top_left"
hover = HoverTool()
hover.tooltips = [("Город", "@Город"), ("Модель", "@Модель"), ("Нижняя граница", "@Нижняя граница"), ("Верхняя граница", "@Верхняя граница")]
p1.add_tools(hover)
deviation = df['fact'] - df['Модель']
source2 = ColumnDataSource(data=dict(Город=df['Город'], deviation=deviation))
p2 = figure(x_range=df['Город'], title="Отклонение фактического населения от прогнозного", plot_height=400, toolbar_location=None)
p2.vbar(x='Город', top='deviation', width=0.4, source=source2, color="red")
p2.xaxis.major_label_orientation = "vertical"
p2.yaxis.axis_label = "Отклонение"
dashboard = column(p1, p2)
show(dashboard)
```
